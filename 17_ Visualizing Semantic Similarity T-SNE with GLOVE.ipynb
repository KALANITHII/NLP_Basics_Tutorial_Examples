"Exploring Semantic Similarity with T-SNE: Visualizing GLOVE Vectors"

This notebook dives into the world of semantic similarity using T-SNE (t-Distributed Stochastic Neighbor Embedding) to visualize GLOVE vectors. The program identifies similar vectors in a given corpus, providing an insightful representation through plots. By leveraging the power of T-SNE and GLOVE, this exploration offers a captivating perspective on the semantic relationships embedded in the vector space, demonstrating the potential of these techniques for understanding and visualizing word similarities in natural language processing tasks.

```python
import numpy as np
```

```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Load GloVe embeddings
embedding_file = 'glove/glove.6B.100d.txt'
embeddings = {}
with open(embedding_file, 'r', encoding='utf8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.array([float(val) for val in values[1:]])
        embeddings[word] = vector

# Find similar vectors using T-SNE
words = ['movie', 'director', 'film', 'camera', 'lion', 'tiger', 'cheetah', 'language', 'speech', 'message']
word_vectors = np.array([embeddings[word] for word in words])
tsne = TSNE(n_components=2, random_state=42, perplexity=5)
word_vectors_2d = tsne.fit_transform(word_vectors)

# Plot the results
plt.figure(figsize=(10, 8))
plt.scatter(word_vectors_2d[:, 0], word_vectors_2d[:, 1])
for i, word in enumerate(words):
    plt.annotate(word, xy=(word_vectors_2d[i, 0], word_vectors_2d[i, 1]))
plt.show()
```
